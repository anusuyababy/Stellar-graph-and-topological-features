# CALCULATING DISTANCE MEASURE USING COSINE DISTANCE

```python
mainDataDFWoClass=mainDataDF.copy()
mainDataDFWoClass.drop(columns=['Class'],inplace=True)
mainDataDFWoClass
```

```python
def self_cosine_vectorized(a):
    dots = a.dot(a.T)
    sqrt_sums = np.sqrt((a**2).sum(1))
    cosine_dists = 1 - (dots/sqrt_sums)/sqrt_sums[:,None]
    np.fill_diagonal(cosine_dists,0)
    return cosine_dists
```

```python
dist = self_cosine_vectorized(mainDataDFWoClass.values) 
dist
```

```python
# Converting array into dataframe format
X1 = pd.DataFrame(dist)
X1.head(10)
```

```python
# taking average
y = np.average(X1)
y
```

```python
adjmat_df=X1.copy()
```

```python
adjmat_df[adjmat_df < y]=0
adjmat_df[adjmat_df >= y]=1
```

### TOPOLOGICAL FEATURE EXTRACTION

```python
in_degreeDict = dict(diGraph.in_degree())

in_degreeDict = collections.OrderedDict(sorted(in_degreeDict.items()))
inDegreeList=list(in_degreeDict.values())
```

```python
out_degreeDict = dict(diGraph.out_degree())
out_degreeDict=collections.OrderedDict(sorted(out_degreeDict.items()))
outDegreeList=list(out_degreeDict.values())
```

```python
hubDict,authDict = nx.hits(graphNw)
hubList=list(hubDict.values())
authList=list(authDict.values())
```

```python
from networkx.algorithms import community
communities = community.greedy_modularity_communities(graphNw)
```

```python
pip install tqdm
```

```python
from tqdm import tqdm
modularity_dict = {} # Create a blank dictionary
for i,c in tqdm(enumerate(communities)): # Loop through the list of communities, keeping track of the number for the community
    for name in c: # Loop through each person in a community
        modularity_dict[int(name)] = i
modularityDict=collections.OrderedDict(sorted(modularity_dict.items()))
modularityList=list(modularityDict.values())
```

```python
PageRankDict=nx.pagerank(graphNw, alpha=0.85)
PageRankList=list(PageRankDict.values())
```

```python
triangleList=[]
for node in tqdm(graphNw.nodes()):
    triangleList.append(nx.triangles(graphNw, node))
#triangleList
```

```python
closenessDict=nx.closeness_centrality(graphNw)
closenessList=list(closenessDict.values())
```

```python
betweennessCentralityDict = nx.betweenness_centrality(graphNw)
betweennessCentralityList=list(betweennessCentralityDict.values())
```

```python
harmonicCentralityDict=nx.harmonic_centrality(graphNw)
harmonicCentralityList=list(harmonicCentralityDict.values())
```

```python
eigenvectorDict = nx.eigenvector_centrality(graphNw) 
eigenvectorList=list(eigenvectorDict.values())
```

```python
graphFeatures_dict = {'Nodes' : nodeList ,'InDegree':inDegreeList,'OutDegree':outDegreeList,'Degree':degreesList, 'WeightedInDegree':inDegreeList,
              'ClosenessCentrality':closenessList,'HarmonicCentrality':harmonicCentralityList,
              'BetweennessCentrality' :betweennessCentralityList,'Authority':authList,'hub':hubList,'Modularity':modularityList,
              'PageRank':PageRankList,'Triangles':triangleList,'EigenvectorCentrality':eigenvectorList,'Class':classList}
graphFeaturesDf = pd.DataFrame.from_dict(graphFeatures_dict)
graphFeaturesDf
```

```python
graphFeaturesWOClass=graphFeaturesDf.copy()
graphFeaturesWOClass.drop(columns=['Class','Nodes'],inplace=True)
scaler = MinMaxScaler()
scaledGraphFeatures = pd.DataFrame(scaler.fit_transform(graphFeaturesWOClass.values), columns=graphFeaturesWOClass.columns, index=graphFeaturesWOClass.index)
scaledGraphFeatures
```

###  TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7, stratify=y)
```

### BASELINE MODEL

```python
from sklearn.linear_model import PassiveAggressiveClassifier
pa=PassiveAggressiveClassifier(max_iter=1000, random_state=7, tol=1e-3)
pa.fit(x_train, y_train)
y_pred=pa.predict(x_test)
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print('accuracy_score: ', accuracy_score(y_test, y_pred))
```

```python
from sklearn.svm import LinearSVC
lsvm = LinearSVC()
lsvm.fit(x_train, y_train)
y_pred=lsvm.predict(x_test)
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print('accuracy_score: ', accuracy_score(y_test, y_pred))
```
